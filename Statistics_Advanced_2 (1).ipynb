## Question 1: What is hypothesis testing in statistics?

**Answer:**  
Hypothesis testing is a statistical method used to determine if there is enough evidence in a sample to infer that a certain condition is true for the entire population.  
It involves:
1. Formulating hypotheses (H₀ and H₁).  
2. Choosing a significance level (α).  
3. Calculating a test statistic.  
4. Making a decision to reject or fail to reject H₀.  

## Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?

**Answer:**  
- **Null Hypothesis (H₀):** Assumes no difference or effect (default position). Example: μ = 50.  
- **Alternative Hypothesis (H₁):** Opposes H₀, claims difference exists. Example: μ ≠ 50.  

**Key Difference:**  
H₀ is conservative baseline (status quo), while H₁ is the claim being tested.  

## Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

**Answer:**  
The **significance level (α)** defines the threshold for rejecting H₀.  
- Common α: 0.05, 0.01.  
- If **p ≤ α** → Reject H₀.  
- If **p > α** → Fail to reject H₀.  

It reflects our tolerance for Type I error (false positive).  


## Question 4: What are Type I and Type II errors? Give examples.

**Answer:**  
- **Type I Error (False Positive):** Rejecting H₀ when it is true.  
  Example: Concluding medicine works, but it actually does not.  

- **Type II Error (False Negative):** Failing to reject H₀ when H₁ is true.  
  Example: Concluding medicine does not work, when in reality it is effective.  

## Question 5: What is the difference between a Z-test and a T-test? When to use each?

**Answer:**  
- **Z-test:** Used when population variance is known or n > 30, relies on Normal distribution.  
- **T-test:** Used when population variance is unknown and sample size is small (n ≤ 30), relies on Student’s t-distribution.  

**Use:**  
- Large samples / known σ → Z-test.  
- Small samples / unknown σ → T-test.

```python
# Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show Python code and interpret the results.

import numpy as np
from scipy import stats

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

mu0 = 50  # hypothesis mean

sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
n = len(sample_data)

# Z-statistic
z = (sample_mean - mu0) / (sample_std / np.sqrt(n))

# p-value (two-tailed)
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

print("Sample Mean:", sample_mean)
print("Z-statistic:", z)
print("P-value:", p_value)

if p_value < 0.05:
    print("Reject H0 → Evidence mean ≠ 50")
else:
    print("Fail to reject H0 → No evidence mean differs from 50")
```

```python
# Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

np.random.seed(0)
data = np.random.normal(loc=50, scale=5, size=1000)

mean = np.mean(data)
se = stats.sem(data)
ci = stats.t.interval(0.95, len(data)-1, loc=mean, scale=se)

print("Sample mean:", mean)
print("95% Confidence Interval:", ci)

# Plot
plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(mean, color='red', linestyle='--', label='Mean')
plt.axvline(ci[0], color='green', linestyle='--', label='95% CI lower')
plt.axvline(ci[1], color='green', linestyle='--', label='95% CI upper')
plt.legend()
plt.title("Normal Distribution with 95% CI")
plt.show()
```

```python
## Question 9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent.

'''**Explanation:**
- Z-scores indicate how many standard deviations a value lies from the mean.
- Z = 0 → at the mean.
- Positive Z → above mean.
- Negative Z → below mean. '''

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Generate dataset
data = np.random.normal(100, 15, 200)

# Compute Z-scores
z_scores = stats.zscore(data)
print("First 10 Z-scores:", z_scores[:10])

# Plot histogram
plt.hist(z_scores, bins=30, color='lightcoral', edgecolor='black', alpha=0.7)
plt.title("Histogram of Z-scores")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.show()
```
